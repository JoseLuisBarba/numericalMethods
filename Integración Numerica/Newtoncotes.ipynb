```python
def interpolacion_lagrange(cls,tabla,min=None,max=None):
  n = len(tabla)
  x = sym.Symbol('x')
  MAX = max
  MIN = min
  Pi = 0
  for i in range(0,n):
    #para lagrange 
    numerador = 1
    denominador = 1
    for j in range(0,n):
      if(j != i):
        numerador = numerador * (x - tabla['valores X'][j])
        denominador = denominador * (tabla['valores X'][i] - tabla['valores X'][j])
    Li = numerador/denominador

    Pi = Pi + Li*tabla['valores Y'][i]
    return Pi
```

```python
def function(cls,funcion,numero):
  """
  Convierte expresiones tipo string en funciones
  """
  x = sympy.symbols('x')
  expresion = funcion
  return (sympy.sympify(expresion).subs(x,numero))
```
