#**MÉTODO DE STEFFENSEN**

> Para encontrar una raíz real de la ecuación g(x) = x, porporcionar la función g(x) y  X0 (valor inicial).

#**Implementación del método:**

*   Precione play en cada bloque descendiendo y esperar que carguen cada uno.

```python
import sympy
import pandas as pd
import numpy as np
from math import *

class Steffensen:
    def __init__(self,funcion,aproximacion_inicial,tolerancia = 1e-6, iteraciones = 100):
        """
        Parameters
        ----------
        funcion : string con variable x
            funcion x = g(x) o f(x) = 0 despejando un x.
        aproximacion_inicial : float
            x aproximado
        tolerancia : float, optional
            DESCRIPTION. The default is 1e-6.
        iteraciones : int, optional
            DESCRIPTION. The default is 100.

        Returns
        -------
        None.

        """
        self.funcion = funcion
        self.P0 = aproximacion_inicial
        self.tol = tolerancia
        self.iteraciones = iteraciones
        self.tabla = None
        self.xvalues = []
    def __function(self,numero):
        x = sympy.symbols('x')
        expresion = self.funcion
        return (sympy.sympify(expresion).subs(x,numero))
    def __Method_Steffensen(self):
        p0 = self.P0
        TOL = self.tol
        i = 1
        while i <= self.iteraciones:
            #PASO 3
            p1 = self.__function(p0)
            p2 = self.__function(p1)
            p = p0 - (((p1 - p0)**2)/(p2 - 2*p1 + p0))
            self.xvalues.append(p)
            #PASO 4
            if (abs(p-p0) < TOL):
                return p
            #PASO 5
            i += 1
            #PASO 6
            p0 = p
        return('El método falló despues de',self.iteraciones,'.')
    def execute(self):
        resultado = self.__Method_Steffensen()
        tabla = pd.DataFrame(
            {
                'x': self.xvalues,
            }
            )
        return(resultado,tabla.head())
      
```
