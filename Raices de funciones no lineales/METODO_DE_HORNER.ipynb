##**Implementación**



1.   Ingrese el grado del polinomio. (El grado debe ser un entero)
2.   Ingrese dentro de las comillas doble los coeficientes del polinomio sin usar espacio y separados por coma.

 

> Ejemplo:

grado = 3
coeficientes = "2,-4,5,7"
Si falta una x^n completar su coeficiente con 0

```python
grado =  3#@param {type:"integer"}
```

```python
coeficientes =  "3,0,-4,-1"#@param {type:"raw"}
coeficientes = coeficientes.split(',')
coeficientes = [float(x) for x in coeficientes]
print(coeficientes)
```

Ingrese el x a evaluar:

```python
t =  2#@param {type:"number"}
```

```python
n = grado

a = coeficientes
y = a[0]
z = a[0]

for i in range(1,n):
  y = t * y + a[i]
  z = t * z + y
y = t * y + a[n]

```

**EL valor de la imagen es:**

```python
print(f'F({t})=',y)
```

**El valor de la derivada es:**

```python
print(f'Df({t})=',z)
```
