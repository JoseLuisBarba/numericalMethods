```python
import matplotlib.pyplot as plt
import pandas as pd
from math import *
import sympy

def function(expresion,numero):
  x = sympy.symbols('x')
  return (sympy.sympify(expresion).subs(x,numero))



def derivate(expresion,point):
    dt = 1e-6
    lim = (function(expresion,point + dt) - function(expresion,point) ) / dt
    return lim


    
def NewtonMethod(expresion,approximation,tolerance=1e-6,iterations=100):
    p0 = approximation
    n = 100
    p = None
    i = 1 # paso 1
    valuesP = []
    while i <= n: #paso 2 -> se hace el paso 3-6
        p = p0 - function(expresion,p0)/derivate(expresion,p0) #paso 3 calculo de pi
        valuesP.append(p)
        if abs(p-p0) < tolerance: 
            values = pd.DataFrame(
            {
            'x': valuesP,
            'y': [j for j in valuesP]
            }
            )
            return (p,values) # paso 4:  el procedimiento fue exitoso
        i += 1 # paso 5 : seguimos buscando
        p0 = p #paso 6 : cambiaos
    print('El método fallo despues de '+ str(n) + ' iteraciones')
```
