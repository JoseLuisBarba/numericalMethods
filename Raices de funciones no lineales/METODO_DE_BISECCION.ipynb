```python
import matplotlib.pyplot as plt
import pandas as pd
from math import *
import sympy



class biseccion:
    def __init__(self,expresion,a,b):
        self.expresion = expresion
        self.p = None
        self.a = a
        self.b = b
        self.valoresP = []
        self.valoresFP = []
        self.FP = 0
        self.FA = 0
    def imagen(self,numero):
        x = sympy.symbols('x')
        return (sympy.sympify(self.expresion).subs(x,numero))
    def tabla(self,xi,yi):
        t = pd.DataFrame(
                {
                    'x':xi,
                    'f(x)':yi
                }
            )
        return t
    def ejecutar(self):
        i = 1
        self.FA = self.imagen(self.a)
        while i <= 100:
            self.p = self.a + (self.b-self.a)/2
            self.valoresP.append(self.p)
            if (self.FP == 0) or (((self.b-self.a)/2)<1e-10):
                self.FP = self.imagen(self.p)
                self.valoresFP.append(self.FP)
                
            i = i + 1
            if (self.FA*self.FP)>0:
                self.a = self.p
                self.FA = self.FP
            else:
                self.b = self.p
        tab = self.tabla( self.valoresP, self.valoresFP)
        return tab   
```
