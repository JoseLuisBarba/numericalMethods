```python

import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

class spline():
  def __init__(self,a,b,c,d,pnti,pntf):
    self.a = a
    self.b = b
    self.c = c
    self.d = d
    self.pnti = pnti
    self.pntf = pntf
  def funcion_str_(self):
    return "self.d*((x-self.pnti)**3)+self.c*((x-self.pnti)**2)+self.b*(x-self.pnti)+self.a"
  def funcion(self,x):
    return self.d*((x-self.pnti)**3)+self.c*((x-self.pnti)**2)+self.b*(x-self.pnti)+self.a
  def __str__(self):
    return f"{self.d}*((x-{self.pnti})**3)+{self.c}*((x-{self.pnti})**2)+{self.b}*(x-{self.pnti})+{self.a}, {self.pnti}<=x<={self.pntf}"



class interpo_Spline():
  def __init__(self,x_valores,y_valores):
    self.__xi = x_valores
    self.__xj = y_valores
    self.cs = CubicSpline(x_valores,y_valores,bc_type="natural")
    self.__splines = []
  def ingresar(self,position,siguiente):
    s = spline(self.cs.c.item(3,position), self.cs.c.item(2,position), self.cs.c.item(1,position), self.cs.c.item(0,position),self.__xi[position],self.__xi[siguiente])
    self.__splines.append(s)
  def crear_funciones(self):
    for i in range(0,len(self.__xi)-1):
      self.ingresar(i,i+1)
  def imprimir_funciones(self):
    print("Las funciones spline son las siguientes: ")
    for i in self.__splines:
      print(i)
    print("")
  def graficar(self):
    plt.figure(figsize=(30,12))
    for i in self.__splines:
      dom = np.linspace(i.pnti, i.pntf, 1000)
      imagen = [i.funcion(j) for j in dom]
      plt.plot(dom,imagen)
    plt.axhline(0)
    plt.axvline(0)
    plt.grid()
    plt.axhline(0)
    plt.axvline(0)
    plt.show()


#@title Ingresar valores de X e Y

if __name__ == '__main__':
  x_values = "-5.25,-5.20,-5,-4.4,-4,-3.2,-3,-2.6,-2.2,-2,-1.4,-1,0,0.6,1,1.2,2,2.4,3,4,4.6,5,5.2" #@param {type:"string"}
  y_values = "0,-0.4,-0.8,-1.4,-1.6,-1.8,-1.75,-1.7,-1.6,-1.55,-1.6,-1.65,-2,-2.6,-2.8,-3,-3.2,-3.2,-3,-2.4,-1.6,-0.4,0" #@param {type:"string"}
  x_values = [float(i) for i in x_values.split(',')]
  y_values = [float(i) for i in y_values.split(',')]
  try:
    u = interpo_Spline(x_values,y_values)
    u.crear_funciones()
    print("Resultados:")
    u.imprimir_funciones()
    u.graficar()
  except Exception as e:
    print("Ocurrió un error tipo ",type(e),": ",e)



```
